```python
import random
import pandas as pd
import jieba

from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from imageio import imread

import matplotlib.pyplot as plt

from palettable.colorbrewer.diverging import Spectral_10
```

```python
font_path = 'ПУТЬ К ФАЙЛУ СО ШРИФТОМ'
#font_path = '/content/KNBobohei-Bold.ttf' ПРИМЕР С ПРОШЛОГО ГОДА

def color_func(word, font_size, position, orientation, random_state=None, **kwargs): #создаем рандом по раскрашивалке цветов
    return tuple(Spectral_10.colors[random.randint(0,9)])

data = pd.read_excel("НАЗВАНИЕ ФАЙЛА")

def get_wordcloud(data, title):
    words = []
    for col in data.columns:
        words.extend(list(data[col]))

    text = ' '.join(words)

    wc = WordCloud(width=1000, height=600, font_path=font_path, background_color="white", max_words=2000, max_font_size=1500, random_state=1)
    wc.generate_from_text(text)
    wc.recolor(color_func=color_func, random_state=2)

    output_path = f'{title}.png'
    wc.to_file(output_path)


    # ВЫВОДИМ ОБЛАКО
    plt.rcParams["figure.figsize"] = (20,20)
    plt.imshow(wc)
    plt.axis("off")
    plt.show()
```

```python
get_wordcloud(data.iloc[:, 2:5], 'whole_wordcloud')
```
