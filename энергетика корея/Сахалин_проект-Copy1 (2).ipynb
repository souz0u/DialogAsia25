Какова корреляция между климатическими условиями и объемами выработки гидроэнергии и геотермальной/солнечной энергии по месяцам в Республике Корея?

```python
df_climate = pd.read_csv('климат.csv')
df_climate
```

Сделаем срез для удобства составления графика

```python
df_climate_1 = df_climate.iloc[ :, 3:]
df_climate_1
```

Транспонируем таблицу для удобства составления графика

```python
rus_labels = ["Гидроэнергетика", "Геотермальная, солнечная энергетика"]
df_climate_1 = df_climate_1.T 
df_climate_1.columns = ["Гидроэнергетика", "Геотермальная, солнечная энергетика"]
df_climate_1
```

```python
clean_index = df_climate_1.index.get_level_values(0).astype(str).str.replace(' Month', '').str.replace(' ', '')
df_climate_1.index = clean_index

ax = df_climate_1.plot(kind = 'line', 
                       linewidth = 2, 
                       marker = 'o', 
                       markersize = 5, 
                       figsize = (16, 8), 
                       grid = True, 
                       rot = 90)

# Определяем, как часто мы хотим видеть метки на оси (например, каждую вторую)
tick_spacing = 2

# Создаем массив ЧИСЛОВЫХ ПОЗИЦИЙ для наших меток.
# np.arange(start, stop, step) создаст последовательность: 0, 6, 12, 18, ...
tick_positions = np.arange(0, len(df_climate_1.index), tick_spacing)

# Получаем ТЕКСТОВЫЕ ПОДПИСИ, которые соответствуют этим числовым позициям.
tick_labels = df_climate_1.index[tick_positions]

# Применяем ПОЗИЦИИ и ПОДПИСИ к осям графика.
ax.set_xticks(tick_positions)
ax.set_xticklabels(tick_labels)

ax.legend(labels = rus_labels, loc = 'upper center', bbox_to_anchor = (0.5, -0.25), ncol = 2)

```

В соответствии с полученными данными можно сделать следующие выводы: пики объемов выработки гидроэнергии на протяжении 10 лет приходятся на июль, что обусловлено сезоном дождей в этот период в Республике Корея, когда выпадает наибольшее количество осадков за год. Увеличение количества осадков положительно влияет на выработку гидроэнергии. 
Также можно отметить, что объемы выработки солнечной энергии в ноябре снижаются. Причиной этого может быть сокращение светового дня до 10 часов, что снижает рассматриваемый показатель. Пик объемов выработки солнечной энергии приходится на апрель, что может быть обусловлено наибольшей активностью солнца в рассматриваемый период.

Выгрузим метерологические данные и данные объемов гидроэнергетики и геотермальной/солнечной энергетики

```python
df_merged = pd.read_csv('data_merged.csv', index_col = 0)
df_merged = df_merged.rename(columns = {'Геотермальная, солнечная энергетика и т.д.':'Геотермальная, солнечная энергетика'})
df_merged
```

Проведем корреляционный анализ

Проверим наличие связи между количеством осадков и объемами гидроэнергетики

```python
hydro_series = df_merged['Гидроэнергетика']
precip_series = df_merged['precip']
correlation_coefficient = hydro_series.corr(precip_series)
print(correlation_coefficient)

cloudcover_series = df_merged['cloudcover']
correlation_coefficient_1 = hydro_series.corr(cloudcover_series)
print(correlation_coefficient_1)
```

Умеренная корреляция. Может свидетельствовать о наличии умеренной связи между двумя параметрами

Проверим взаимосвязь между параметром пасмурности неба и объемами гидроэнергетики

```python
cloudcover_series = df_merged['cloudcover']

# Используем метод .corr() для расчета коэффициента корреляции между ними
correlation_coefficient_1 = hydro_series.corr(cloudcover_series)

correlation_coefficient_1
```

Высокая корреляция Пирсона, может свидетельствовать о взаимосвязи двух факторов (пасмурное небо - увеличение объемов гидроэнергетики)

Проведем корреляционный анализ для параметров "объемы геотермальной и солнечной энергетики" и "солнечная радиация"

```python
solar_series = df_merged['Геотермальная, солнечная энергетика']
solarradiation_series = df_merged['solarradiation']

# Используем метод .corr() для расчета коэффициента корреляции между ними
correlation_coefficient_2 = solar_series.corr(solarradiation_series)

correlation_coefficient_2
```

Умеренная корреляция между двумя параметрами. Может свидетельствовать о наличии определенной связи (сильная солнечная радиация - увеличение объемов солнечной энергетики).
